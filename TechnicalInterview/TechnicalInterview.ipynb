```python
class Classy(object):
    def __init__(self):
        self.items = []
        
    def addItem(self, input):
        self.items.append(input)
        
    def getClassiness(self):
        classiness = 0
        sum_dict = {'tophat': 2, 'bowtie': 4, 'monocle': 5}
        for it in self.items:
            value = sum_dict.get(it)
            classiness = classiness + value if value else classiness
        return classiness

me = Classy()
print 'first', me.getClassiness()
me.addItem('tophat')
print 'second', me.getClassiness()
me.addItem("bowtie")
me.addItem("jacket")
me.addItem("monocle")
print 'third', me.getClassiness()
me.addItem("bowtie")
print 'fourth', me.getClassiness()
```

```python
# Linked Lists

class Element(object):
    def __init__(self, value):
        self.value = value
        self.next = None
        
class LinkedList(object):
    def __init__(self, head=None):
        self.head = head
        
    def append(self, new_element):
        current = self.head
        if self.head:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            self.head = new_element
            
    def get_position(self, position):
        """Get an element from a particular position.
        Assume the first position is "1".
        Return "None" if position is not in the list."""
        current = self.head
        i = 1
        if position < 1:
            return None
        while i <= position:
            if i == position:
                return current
            if not current.next:
                break
            current = current.next
            i += 1
        return None

    def insert(self, new_element, position):
        """Insert a new node at the given position.
        Assume the first position is "1".
        Inserting at position 3 means between
        the 2nd and 3rd elements."""
        current = self.head
        successor = current.next
        if position < 1:
            return None
        if position == 1:
            new_element.next = self.head
            self.head = new_element
            return None
        i = 1
        while i < position - 1:
            current = current.next
            if current is None:
                return None
            successor = current.next
            i += 1
        current.next = new_element
        new_element.next = successor
        return
    
    def delete(self, value):
        """Delete the first node with a given value."""
        current = self.head
        if current.value == value:
            self.head = current.next
            return
        else:
            while current.next:
                if current.next.value == value:
                    successor = current.next.next
                    current.next = successor
                    return
        
    
# Test cases
# Set up some Elements
e1 = Element(1)
e2 = Element(2)
e3 = Element(3)
e4 = Element(4)

# Start setting up a LinkedList
ll = LinkedList(e1)
ll.append(e2)
ll.append(e3)

# Test get_position
# Should print 3
print '3', ll.head.next.next.value
print ll.get_position(3).value

# Test insert
ll.insert(e4,3)
# Should print 4 now
print '4', ll.get_position(3).value

# Test delete
ll.delete(1)
# Should print 2 now
print '2', ll.get_position(1).value
# Should print 4 now
print '4', ll.get_position(2).value
# Should print 3 now
print '3', ll.get_position(3).value
```

```python
# Queue

class Queue:
    def __init__(self, head=None):
        self.storage = [head]

    def enqueue(self, new_element):
        return self.storage.append(new_element)

    def peek(self):
        return self.storage[0] if self.storage else None

    def dequeue(self):
        return self.storage.pop(0) if self.storage else None
    
# Setup
q = Queue(1)
q.enqueue(2)
q.enqueue(3)

# Test peek
# Should be 1
print '1', q.peek()

# Test dequeue
# Should be 1
print '1', q.dequeue()

# Test enqueue
q.enqueue(4)
# Should be 2
print '2', q.dequeue()
# Should be 3
print '3', q.dequeue()
# Should be 4
print '4', q.dequeue()
q.enqueue(5)
# Should be 5
print '5', q.peek()
```

```python
# Binary Search

def binary_search(input_array, value):
    """Your code goes here."""
    low = 0
    high = len(input_array) - 1
    while low <= high:
        mid = (low + high) / 2
        if input_array[mid] == value:
            return mid
        elif input_array[mid] < value:
            low = mid + 1
        else:
            high = mid - 1
    return -1

test_list = [1,3,9,11,15,19,29]
test_val1 = 25
test_val2 = 15
print '25', binary_search(test_list, test_val1)
print '15', binary_search(test_list, test_val2)
```

```python
# Recursion
# Fibonacci sequence

def get_fib(position):
    if position < 0:
        return -1
    if position == 0:
        return 0
    if position == 1:
        return 1
    return (get_fib(position - 1) + get_fib(position - 2))

# Test cases
print get_fib(9)
print get_fib(11)
print get_fib(0)
```

```python
# Hash functions and Strings:

class HashTable(object):
    def __init__(self):
        self.table = [None]*10000

    def store(self, string):
        """Input a string that's stored in 
        the table."""
        index = self.calculate_hash_value(string)
        if index == -1:
            return
        if self.table[index] != None:
            self.table[index].append(string)
        else:
            self.table[index] = [string]

    def lookup(self, string):
        """Return the hash value if the
        string is already in the table.
        Return -1 otherwise."""
        index = self.calculate_hash_value(string)
        if index == -1 or self.table[index] is None:
            return -1
        return index if string in self.table[index] else -1

    def calculate_hash_value(self, string):
        """Helper function to calulate a
        hash value from a string."""
        if len(string) < 2:
            return -1
        hash_value = ord(string[0]) * 100 + ord(string[1])
        return hash_value
    
# Setup
hash_table = HashTable()

# Test calculate_hash_value
# Should be 8568
print hash_table.calculate_hash_value('UDACITY')

# Test lookup edge case
# Should be -1
print hash_table.lookup('UDACITY')

# Test store
hash_table.store('UDACITY')
# Should be 8568
print hash_table.lookup('UDACITY')

# Test store edge case
hash_table.store('UDACIOUS')
# Should be 8568
print hash_table.lookup('UDACIOUS')
print 
    
```
